## ALGORITMO DE SHOR

#### Introducción

En los años 90 se sabia que un computador cuantico podria ser mas eficiente que cualquier computador clasico para ciertos problemas de complejidad computacional. Las investigaciones en esas areas estaban principalmente motivadas por curiosidad academica y no habia mucho motivo economico para invertir en el desarrollo de un computador cuantico.

Eso cambió en 1995 cuando peter shor, descubrió un algoritmo en tiempo cercano al polinomial para factorizar grandes numeros en sus factores primos.

A pesar de que cualquier entero tiene una descomposición unica en un producto de numeros primos, encontrar los factores es considerado un problema hard, de hecho, la seguridad de muchas transacciones se fian de que factorizar enteros con mil digitos o más es practicamente imposible, siendo esto ultimo retado por el algoritmo que encontró peter shor.

#### Algoritmo

Para empezar tenemos el valor N que es el que queremos factorizar, debemos considerar:

1. Escogemos un a aleatorio con a<N tal que a y N sean coprimos, es decir que el **gcd(a,N)=1**.  
Entonces es posible encontrar un entero $r\in[1,N]$ tal que:  
$a^r\ mod\ N = 1$  
donde *r* es el orden de *a* en el modulo N  

2. Para $a$ con orden $r\ mod\ N$, la función de exponenciación modular  
$f(x)=a^x\ mod\ N$  
es periodica con periodo $r$ si  
    $f(x+r) = f(x) \\ f(x+r) = a^{x+r}\ mod\ N \\ = (a^x\ mod\ N)(a^r\ mod\ N) \\ = (a^x\ mod\ N)*(1) = f(x)$  
   por lo tanto encontrar el periodo es lo mimo que encontrar el orden de algún coprimo de N

3. Si encontramos entonces que *r* es par, se construye $b=a^{r/2} $ donde  
$ b^2 \equiv 1\ mod\ N \\ b^2 -1 \equiv 0\ mod\ N \\ (b-1)(b+1) \equiv 0 \ mod\ N$  
Mientras $b \neq \pm 1$ entonces al menos $(b-1)(b+1)$ debe tener un factor no trivial en común con N  
Entonces calculando **gcd(b-1,N)** y **gcd(b+1,N)** para obtener un factor de N

Por ejemplo, para N = 15 y k = 7, la secuencia de exponenciación modular es 1, 7, 4, 13, 1, 7, 4, 13, 1, . . . con periodo 4. Como el periodo es par, podemos aplicar la idea anterior:  

$7^4\ mod\ 15 ≡ 1 ⇒ 7^4 − 1\ mod\ 15 ≡ 0 ⇒ (7^2 − 1)(7^2 + 1)\ mod\ 15 ≡ 0\ mod\ 15 ⇒(48)(50)\equiv 0\ mod\ 15 ⇒ 15\ divides\ 48 · 50$,  

que puede ser usada para computar los factores de 15 como **gcd(48, 15)= 3** y **gcd(50, 15) = 5**.

El periodo se puede encontrar en computación cuantica usando la transofrmada quantica de fourier (QFT), En la siguiente sección daremos una definición y describiremos el circuito para computar la QFT.

#### ** Codigo en QISKit del circuito de la figura 9**

```python
from qiskit import ClassicalRegister, QuantumRegister
from qiskit import QuantumCircuit, execute
from qiskit.tools.visualization import plot_histogram, circuit_drawer
from math import pi

q = QuantumRegister(5)
c = ClassicalRegister(3)
qc = QuantumCircuit(q, c)
qc.h(q[0])
qc.h(q[1])
qc.h(q[2])
qc.cx(q[2], q[3])
qc.cx(q[2], q[4])
qc.h(q[1])
qc.cu1(pi/2, q[1],q[0]) #theta, control, target
qc.h(q[0])
qc.cu1(pi/2, q[1],q[2]) #theta, control, target
qc.cu1(pi/4, q[0],q[2]) #theta, control, target
qc.h(q[2])
for i in range(3):
    qc.measure(q[i], c[i])

job = execute(qc, backend="local_qasm_simulator", shots=1024, max_credits=3)
plot_histogram(job.result().get_counts(qc))
circuit_drawer(qc)
```

### PASOS DE IMPLEMENTACIÓN - Matthew Hayward

Un pequeño resumen de esta implementación: 
El algoritmo de shor intenta encontrar el periodo r de $x^a \ mod \ N$, (debido a la explicación anterior aqui cambia $a^r \ mod \ N$ por $x^a \ mod \ N$), siendo N el numero a factorizar y siendo $x$ coprimo con N.

Para hacer esto el algoritmo de shor crea un registro de memoria cuantica con 2 partes (puede ser interpretado como 2 registros de qubits). En el primer registro el algoritmo pone en superposición los enteros los cuales serán los $a$ en la función  $x^a \ mod \ N$. Esta implementación escoge los $a$ para ser enteros entre $0$ hasta $q-1$ siendo q una potencia de 2 tal que $n^2 \leq q < 2n^2$. Entonces el algoritmo calcula $x^a \ mod \ N$ donde $a$ es la superposición de estados y pone el resultado en el segundo registro.

A continuación el algoritmo mide el estado del segundo registro, el cual contiene las superposiciónes de todas las posibles salidas para $x^a \ mod \ N$. Medir este registro tiene efecto en colapsar el estado en algun valor $k$. Tambien tiene el efecto de proyectar el primer registro en un estado consistente con la medición del 2do registro. Esto significa, despues de la medición del segundo registro se obtiene un valor k, el primer registro queda con los estados base que si son ingresados en $x^a \ mod \ N$ se obtiene k. Como sabemos que $x^a \ mod \ N$ es una función periodica, sabemos que el registro 1 contendra valores  c, c + r, c + 2r... donce c es el menor entero posible tal que $x^c \ mod \ N = k$

* **Paso 1**

Determine si el numero N es primo, par o potencia entera de un numero primo. Si lo es entonces no use el algoritmo de shor, hay métodos clásicos eficientes para determinar si un entero n pertenece a uno de los grupos mencionados y porveer factores para ellos. Este paso se hace en un computador clásico.

* **Paso 2** (propio de esta implementación)

Escoja un entero q tal que sea potencia de 2 tal que $n^2 \leq q < 2n^2$. Este paso se hace en un computador clásico.

* **Paso 3**

Escoja un entero aleatorio $x$ que sea coprimo con N (**gcd(x,N)=1**). Hay métodos clásicos eficientes para encontrar el x. Este paso se hace en un computador clásico.

* **Paso 4**

Cree 2 registros de qubits, registro 1 y registro 2. De esta forma el estado de nuestro computador cuantico estára dado por $\lvert reg1, reg2\rangle$. El registro 1 debe tener suficientes qubits para representar enteros tan grandes como q-1. El registro 2 debe tener suficientes qubits para representar enteros tan grandes como n-1. Los calculos para saber cuantos qubits son necesarios se harán en un computador clásico.

* **Paso 5**

Cargue el registro 1 con una superposición de todos los enteros desde 0 hasta q-1. Cargue el registro 2 con todo en 0. esta operación será realizada en un computador cuantico. El estado dotal del registro de memoria cuantica en este momento es:

${\frac{{1}}{{\sqrt{q}}}} \sum_{{a = 0}}^{{q - 1}}\lvert a, 0 \rangle$

* **Paso 6**

Ahora aplique la transformación $x^a \ mod\ N$ para cada numero guardado en el registro un y almacene el resultado en el registro 2. Debido al paralelismo cuantico esta operación solo toma un paso, como el computador cuantico solo calcula  $ x^{\lvert a\rangle}\ mod\ N$, donde $\lvert a\rangle$ es la superposicion de estados creados en el paso 5. El estado del registro en este paso es:

${\frac{{1}}{{\sqrt{q}}}} \sum_{{a = 0}}^{{q - 1}} \lvert a, x^a\ mod\ N \rangle$

(Aplicar la multiplicación modular explonencial en un circuito cuantico se encuentra en [9] )

* **Paso 7**

Mida el segundo registro y observe algún valor k. Este tiene el efecto de colapsar el registro1 en una superposición equivalente de cada valor de $a$ entre $0$ y $q-1$ como se explicó al inicio. El estado del registro en este paso es:

$\frac{{1}}{{\sqrt{\vert\vert A\vert\vert}}} \sum_{{a'=a'\in A}}^{} \lvert  a', k\rangle$  
Donde A es el conjunto de $a$ tal que $x^a \ mod\ N = k$  y $\lvert\lvert A\lvert\lvert $ es el numero de elementos en ese conjunto

* **Paso 8**

Ahora aplique la transformada discreta de fourier en el registro 1.

* **Paso 9**

Mida el estado del registro 1, llame este valor $m$, este entero $m$ tiene una alta probabilidad de ser multiplo de $\frac{q}{r}$, donde $r$ es el periodo deseado. Este paso es realizado por el computador cuántico.

* **Paso 10**

Tome el valor m, y en el computador clásico haga post-procesado para calcular $r$ basado en en el conocimiento de $m$ y $q$. En particular:
   * $m$ tiene alta probabilidad de ser = $ \lambda$*(q/r) where $ \lambda$ is an integer
   * Si realizamos una división de punto flotante de $\dfrac{m}{q}$, y calculamos la mejor aproximación racional cuyo denominador es menor o igual a $q$, podemos tomar este valor como candidato para $r$
   * Si nuestro candidato para $r$ es impar lo duplicamos si al hacerlo obtenemos un valor menor que $q$

Hay formas eficientes de hacer este procesamiento en el computador clásico.


* **Paso 11**

Una vez obtenido $r$, un factor de $n$ puede ser determinado tomando $gcd(x^{r/2} + 1, n)$ y $gcd(x^{r/2} - 1, n)$. Si se encontró un factor de n, termine, caso contrario regrese al paso 4. Este paso final se hace en el computador clásico.
