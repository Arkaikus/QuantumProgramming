## Algoritmo de Busqueda de Grover

### Introducción

Suponga que tiene una lista de N items. Entre estos items hay uno con una propiedad unica que se desea encontrar. Piense en cada elemento de la lista como una caja de un color particular, digamos que todos los items en la lista tienen color gris mientras que el elemento que estamos buscando tiene color rosa.

Clasicamente buscar un elemento en una lista desordenada tiene una complejidad O(N) donde N es el numero de elementos en la lista. El algoritmo de busqueda de Grover que trabaja en un computador cuantico puede encontrar el elemento que estamos buscando en una complejidad $O(\sqrt{N})$

### El Oraculo

Como se provee la lista de items al computador cuantico? una manera comun es codificar cada elemento en terminos de una función $f(x)=0$ para todos los items no marcados x y $f(w)=1$ para el item que estamos buscando. Para usar un computador cuantico para este problema, debemos proveer los items en superposición a esta función. 

De esta forma convertimos la función en una matriz unitaria llamada oraculo. Primero escogemos una codificación binaria de los items $x,y \in {0,1}^n$ de tal forma que N=2^n y de esta forma se puedan representar en terminos de qubits en un computador cuantico. Entonces definimos la matriz oraculo $U_f$ que actue sobre cualquiera de los estados bases $|x\rangle$ haciendo:

$U_f|x\rangle=(-1)^{f(x)}|x\rangle$

Vemos que si x es un item que no buscamos el oraculo no hace nada al estad. Sin embargo, cuando aplicamos el oraculo al estado base $|w\rangle$ lo que hace es cambiar la fase de este, siendo $U_f|w\rangle= -|w\rangle$.

### Amplificación de Amplitudes

Antes de buscar en la lista de items no se tiene idea donde se encuentra el item marcado. Por lo tanto cualquier escogencia de su ubicacion es tan buena como cualquier otra. lo cual puede ser expresado en terminos de un estado cuantico llamado superposición uniforme

$|s\rangle= \dfrac{1}{\sqrt{N}} \sum_{x=0}^{n-1} |x\rangle$

Si se mide en este punto el estado, este colapsará, y las probabilidades de encontrar el valor esperado es $\dfrac{1}{2^n}$ y en promedio nos tomaría intentar $2^n$ veces para encontrar el valor que buscamos.

El procedimiento llamado amplificación de aplitud, el cual es como un computador cuantico mejora la probabilidad. Este procedimiento amplifica la aplitud del item marcado, y disminuye la aplitud de los otros items, de tal forma que midiendo el estado final retorna el item buscado con buena certeza.

Este algoritmo tiene una buena interpretación geometrica. Se genera una rotación en un espacio bidimentional. Los 2 estados especiales que necesitamos considerar son el que tiene el elemento de busqueda $\lvert w\rangle$ y el de la superposición $\lvert s\rangle$. Estos 2 vectores componen un estado bidimensional. No son necesariamente perpendiculares debido a que $\lvert w\rangle$ tiene una amplitud correspondiente en $\lvert s\rangle$. Sin embargo, se introduce un estado adicional $\lvert s'\rangle$ que es un espacio entre estos dos vectores y es perpendicular. Se consigue desde |$s\rangle$ removiendo $\lvert w\rangle$ y reescalando.

- **Paso 0:** La amplificación empieza con la superposición |$s\rangle$.
- **Paso 1:** Se aplica la reflexión $U_f\lvert x\rangle$
- **Paso 2:** Se aplica una reflexión aditiva $U_s$ sobre $\lvert s\rangle$. en notación bra-ket $U_s=2\lvert s\rangle\langle s\lvert -1$  
  Esto lo que hace es aumentar la amplitud del elemento que buscamos y disminuye la de los demás
- Se repite desde el **Paso 1** un numero $\sqrt{N}$ de veces,

### Codigo en QISKit para encotrar 10 con 2 Qubit

```python
from qiskit import ClassicalRegister, QuantumRegister
from qiskit import QuantumCircuit, execute
from qiskit.tools.visualization import plot_histogram, circuit_drawer

q = QuantumRegister(3)
c = ClassicalRegister(2)
qc = QuantumCircuit(q, c)
#qc.x(q[0])
#qc.x(q[1])
qc.h(q[1])
qc.h(q[2])
#Oracle:
#qc.h(q[0]); qc.cx(q[1],q[0]); qc.tdg(q[0]); 
#qc.cx(q[2],q[0]); qc.t(q[0]); qc.cx(q[1],q[0]); qc.tdg(q[0])
#qc.cx(q[2],q[0]); qc.t(q[0]); qc.tdg(q[1]); qc.h(q[0]); qc.cx(q[2],q[1]);
#qc.tdg(q[1]);qc.cx(q[2],q[1]); qc.s(q[1]); qc.t(q[2])

qc.s(q[2])
qc.h(q[1])
qc.cx(q[2],q[1])
qc.h(q[1])
qc.s(q[2])

#Operador de Grover
qc.h(q[1]); 
#qc.barrier(q[1])
qc.h(q[2]); qc.x(q[1]); qc.x(q[2]); qc.h(q[1]);qc.cx(q[2],q[1])
qc.h(q[1]); qc.x(q[1]); qc.x(q[2]); qc.h(q[1]); qc.h(q[2]);

#qc.measure(q,c)
qc.measure(q[1],c[0])
qc.measure(q[2],c[1])
job_exp = execute(qc, backend="local_qasm_simulator", shots=1024, max_credits=3)
plot_histogram(job_exp.result().get_counts(qc))
print(job_exp.result().get_counts(qc))
circuit_drawer(qc)
```
